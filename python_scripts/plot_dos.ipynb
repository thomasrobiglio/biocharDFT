```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open ('C.dos', 'r')
next(f)
l = []
l = [[float(num) for num in line.split()] for line in f]
```

```python
dos_array = []
E_array = []
for i in range(len(l)):
    dos_array.append(l[i][1])
    E_array.append((l[i][0])-4.8536)
```

```python
plt.plot(E_array, dos_array)
plt.xlabel(r'$E-E_f$  [eV]', fontsize=15)
plt.ylabel(r'DOS [states/eV]', fontsize=15)
plt.ylim(0,6)
plt.xlim(-10,10)
#plt.axvline(x = 45, color = 'r', linestyle='--')
```
