## TUTORIAL 1 - Convergence tests

```python
import matplotlib.pyplot as plt
```

```python
f = open ('energy.txt', 'r')
l = []
l = [[float(num) for num in line.split()] for line in f]
print(l)
```

```python
l[1][1]
```

```python
atomic_energy_array = []
e_cut_array = []
for i in range(len(l)):
    atomic_energy_array.append(l[i][1]*13.6/4)
    e_cut_array.append((l[i][0]))
```

```python
plt.plot(e_cut_array, atomic_energy_array)
plt.xlabel(r'$E_\mathrm{{cut}}$ [Ry]', fontsize=15)
plt.ylabel(r'Energy/atom [eV]', fontsize=15)
plt.axvline(x = 45, color = 'r', linestyle='--')
```

```python
g = open ('energy_kpoints.txt', 'r')
m = []
m = [[float(num) for num in line.split()] for line in g]
print(m)
```

```python
atomic_energy_array = []
k_points_array = []
for i in range(len(m)):
    atomic_energy_array.append(m[i][1]*13.6/4)
    k_points_array.append((m[i][0]))
```
