```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_grid = [6, 12, 18, 24, 30, 36, 42, 48]
```

```python
for i in range(len(k_grid)):
    filename = 'C_k'+str(k_grid[i])+'.dos'
    
    f = open (filename, 'r')
    next(f)
    l = []
    l = [[float(num) for num in line.split()] for line in f]
    
    dos_array = []
    E_array = []
    for k in range(len(l)):
        dos_array.append(l[k][1])
        E_array.append((l[k][0])-4.8536)
    plt.plot(E_array, dos_array, label=r'$k=$'+str(k_grid[i]))
    plt.xlabel(r'$E-E_f$  [eV]', fontsize=15)
    plt.ylabel(r'DOS [states/eV]', fontsize=15)
    plt.ylim(0,12)
    plt.legend()
```
